## Deel 1

```python
# Read the lines
day_1_input = open('inputs/dag1').read().splitlines()
day_1_input = [re.split(r"(?<=[R|L])(?=\d)|-", line) for line in day_1_input]

dial_pos = 50
dial_at_zero_counter = 0

def new_dial_position(cur_pos : int, ticks : int, dial_max = 100) : return (cur_pos + ticks) % dial_max 

for direction, ticks in day_1_input:
    ticks = int(ticks)
    delta_dial_pos = ticks if direction == "R" else -ticks
    dial_pos = new_dial_position(dial_pos, delta_dial_pos)
    dial_at_zero_counter += (dial_pos == 0)

print(dial_at_zero_counter)
```

```python
dial_pos = 50
dial_at_zero_counter = 0
for direction, ticks in day_1_input:
    ticks = int(ticks)
    delta_dial_pos = ticks if direction == "R" else -ticks
    full_turns = abs(delta_dial_pos) // 100
    abs_dial_pos = dial_pos + delta_dial_pos

    if abs_dial_pos > 0 : abs_dial_pos -= 100 * full_turns
    if abs_dial_pos < 0 : abs_dial_pos += 100 * full_turns

    dial_at_zero_counter += ((dial_pos != 0) and (abs_dial_pos <= 0 or abs_dial_pos >= 100)) + full_turns
    dial_pos = new_dial_position(dial_pos, delta_dial_pos)

dial_at_zero_counter
```
