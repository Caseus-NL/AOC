## Deel 1

```python
day_1_input = T(parse(open('inputs/dag1').read(), lines, split_numbers))
day_1_test_input = T(parse("""3   4
4   3
2   5
1   3
3   9
3   3""", lines, split_numbers))

def list_distance(left_list : list, right_list : list):
    return sum(abs(a - b) for a, b in zip(left_list, right_list))

test(lambda w: list_distance(sorted(w[0]), sorted(w[1])), day_1_test_input, 11)
run(lambda w: list_distance(sorted(w[0]), sorted(w[1])), day_1_input)
```

## Deel 2

```python
def similarity_score(left_list : list, right_list : list):
    return sum([left * right_list.count(left) for left in left_list])

test(lambda w: similarity_score(w[0], w[1]), day_1_test_input, 31)
run(lambda w: similarity_score(w[0], w[1]), day_1_input)
```
