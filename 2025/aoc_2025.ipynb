# Dag 1 – Secret entrance
We gaan weer los!
De eerste puzzel dit jaar gaat over het invoeren van een soort kluis-code.
We krijgen een set met instructies bestaande uit een letter (R of L) en een getal, die aangeeft hoe we de code in moeten vullen. `L` betekent naar links draaien en het getal erachter geeft aan hoeveel stappen.

## Deel 1
Hoe vaak komen we na het uitvoeren van een instructie <b>precies</b> op 0?

```python
# Read the lines
day_1_input = open('inputs/dag1').read().splitlines()
day_1_input = [re.split(r"(?<=[R|L])(?=\d)|-", line) for line in day_1_input]

dial_pos = 50
dial_at_zero_counter = 0

def new_dial_position(cur_pos : int, ticks : int, dial_max = 100) : return (cur_pos + ticks) % dial_max 

for direction, ticks in day_1_input:
    ticks = int(ticks)
    delta_dial_pos = ticks if direction == "R" else -ticks
    dial_pos = new_dial_position(dial_pos, delta_dial_pos)
    dial_at_zero_counter += (dial_pos == 0)

print(f"Dag 1, deel 1: {dial_at_zero_counter}")
```

## Deel 2
Nu is de vraag hoevaak we het getal 0 <b>passeren</b>, ongeacht of we erop eindigen of niet.

```python
dial_pos = 50
dial_at_zero_counter = 0
for direction, ticks in day_1_input:
    ticks = int(ticks)
    delta_dial_pos = ticks if direction == "R" else -ticks
    full_turns = abs(delta_dial_pos) // 100
    abs_dial_pos = dial_pos + delta_dial_pos

    if abs_dial_pos > 0 : abs_dial_pos -= 100 * full_turns
    if abs_dial_pos < 0 : abs_dial_pos += 100 * full_turns

    dial_at_zero_counter += ((dial_pos != 0) and (abs_dial_pos <= 0 or abs_dial_pos >= 100)) + full_turns
    dial_pos = new_dial_position(dial_pos, delta_dial_pos)

print(f"Dag 1, deel 2: {dial_at_zero_counter}")
```
